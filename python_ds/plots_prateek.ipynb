```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/Users/sunilkumar/Downloads/profile_oxford_v2 - sub_system_profile_oxford_v2.csv')
```

```python
# Plotting a pie chart
plt.figure(figsize=(20, 20))
plt.pie(df['column_count'], labels=df['table_name'], autopct='%1.1f%%', startangle=90, counterclock=False,rotatelabels=True)
 #autopct='%1.1f%%', startangle=90, counterclock=False, rotatelabels=True
plt.title('Distribution of Values for Different Variables')
plt.show()
```

```python
# Plotting with logarithmic scale on y-axis
plt.figure(figsize=(15, 8))
plt.bar( df['table_name'],df['column_count'], color='skyblue')
#plt.bar( df['table_name'],df['sample_row_count'], color='skyblue')
#sample_row_count
plt.xlabel('Variable')
plt.ylabel('Value (Log Scale)')
plt.title('Values for Different Variables (Logarithmic Scale)')
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
